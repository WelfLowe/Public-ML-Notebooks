```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
# Step 1: Generate regression data
np.random.seed(42)
N = 100  # Number of points
x1 = np.random.uniform(-2, 2, N)
x2 = np.random.uniform(-2, 2, N)
y = np.sin(x1) + np.cos(x2) + np.random.normal(0, 0.1, N)  # Some nonlinear function with noise

X = np.column_stack((x1, x2))
```

```python
# Step 2: Train KNN Regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X, y)

# Step 3: Visualization of predictions
x1_range = np.linspace(-2, 2, 1000)
x2_range = np.linspace(-2, 2, 1000)
X1_grid, X2_grid = np.meshgrid(x1_range, x2_range)
X_grid = np.column_stack((X1_grid.ravel(), X2_grid.ravel()))

# Predict over the grid
y_pred = knn.predict(X_grid).reshape(X1_grid.shape)

# Plot data points and prediction regions
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1_grid, X2_grid, y_pred, levels=200, cmap="coolwarm", alpha=0.75)
plt.colorbar(contour, label="Predicted value")
plt.scatter(x1, x2, c=y, edgecolors='k', cmap="coolwarm", label="Data points")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("KNN Predictions")
plt.legend()
plt.show()
```
