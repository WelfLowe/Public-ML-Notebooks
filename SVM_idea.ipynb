```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate x1 values
N = 500  # Number of samples per class
x1_class0 = np.random.uniform(-2, 2, N)  # Random x1 values for class 0
x1_class1 = np.random.uniform(-2, 2, N)  # Random x1 values for class 1

# Generate x2 values based on x1^2 with some noise
x2_class0 = x1_class0**2 - np.random.uniform(0.5, 1.5, N)  # Below the curve
x2_class1 = x1_class1**2 + np.random.uniform(0.5, 1.5, N)  # Above the curve

# Plot the distributions
plt.figure(figsize=(8, 6))
plt.scatter(x1_class0, x2_class0, color="blue", alpha=0.5, label="Class 0 (Below)")
plt.scatter(x1_class1, x2_class1, color="red", alpha=0.5, label="Class 1 (Above)")

# Plot the separation curve x2 = x1^2
x_curve = np.linspace(-2, 2, 100)
y_curve = x_curve**2
plt.plot(x_curve, y_curve, color="black", linestyle="dashed", label="$x_2 = x_1^2$")

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.title("Two-Class Distribution with Quadratic Separation")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Compute x3 = x2 - x1^2
x3_class0 = x2_class0 - x1_class0**2
x3_class1 = x2_class1 - x1_class1**2

# 3D Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1_class0, x2_class0, x3_class0, color="blue", alpha=0.5, label="Class 0 (Below)")
ax.scatter(x1_class1, x2_class1, x3_class1, color="red", alpha=0.5, label="Class 1 (Above)")

ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$x_3 = x_2 - x_1^2$")
ax.set_title("3D Representation with Transformed Feature")
ax.legend()
plt.show()
```
