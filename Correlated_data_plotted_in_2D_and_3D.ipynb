```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(42)
N = 500
r1 = -0.95
r2 = 0.0
mean = [20, 20, 0]
cov_correlated = [[1, r1, r2], [r1, 1, r2], [r2, r2, 0.1]]  # Covariance matrix with correlation r
data_correlated = np.random.multivariate_normal(mean, cov_correlated, N)
data_correlated.shape
```

```python
# Plot the distributions
plt.figure(figsize=(8, 6))
plt.scatter(data_correlated[:,0], data_correlated[:,1], color="blue", alpha=0.5)
plt.title("Correlated 2D data")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(True)

plt.show()
```

```python
# Plot the distributions
plt.figure(figsize=(8, 6))
plt.scatter(data_correlated[:,0], data_correlated[:,2], color="blue", alpha=0.5)
plt.title("Two-Class Distribution with Quadratic Separation")
plt.show()
```

```python
# Plot the distributions
plt.figure(figsize=(8, 6))
plt.scatter(data_correlated[:,1], data_correlated[:,2], color="blue", alpha=0.5)
plt.title("Two-Class Distribution with Quadratic Separation")
plt.show()
```

```python
# 3D Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data_correlated[:,0], data_correlated[:,1], data_correlated[:,2], color="blue", alpha=0.5)

ax.set_title("3D Representation with Transformed Feature")
ax.legend()
plt.show()
```

```python
data2d = data_correlated[:,0:2]
```
